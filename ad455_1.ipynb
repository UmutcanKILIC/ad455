```r
library(data.table)
library(DT)
options(repr.matrix.max.rows=20, repr.matrix.max.cols=30) # for limiting the number of top and bottom rows of tables printed 
```

```r
airport_codes <- data.table::fread('https://datahub.io/core/airport-codes/r/airport-codes.csv')
```

```r
names(airport_codes)
```

```r
airport_codes
```

```r
airport_codes[type == 'small_airport' & elevation_ft > 3000 & iso_country == 'US'] 
# small airports in the US located at an altitude higher that 3000 ft.
```

```r
airport_codes[type == 'small_airport' & elevation_ft > 3000 & iso_country == 'US', .(max_alt = max(elevation_ft), min_alt = min(elevation_ft), avg_alt = mean(elevation_ft))]
# find maximum, minimum, and average altitute for small airports in the US with altitudes higher than 3000 ft.
```

```r
airport_codes[type == 'small_airport' & elevation_ft > 3000 & iso_country == 'US', relative_alt := elevation_ft - mean(elevation_ft)]
# finds the relative altitute compared to the mean altitude for each small airport in the US with elevation higher than 3000 ft.
```

```r
airport_codes[, .(max_alt_countryandtype = max(elevation_ft, na.rm = T), min_alt_countryandtype = min(elevation_ft, na.rm = T)), by = c("iso_country", "type")]
# finds the highest an lowest altitue for each type of airport in each country.
```
